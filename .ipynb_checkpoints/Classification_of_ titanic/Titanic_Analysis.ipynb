# Data Preprocessing

```python
data.isna().sum()
```

```python
# handling Null values
data.loc[data['Age'].isna()]
```

```python
data['Age'].fillna(data['Age'].mean(), inplace=True)
data['Embarked'].fillna(data['Embarked'].mode()[0],inplace=True)
```

```python
# Remove unwanted columns
data.drop('Cabin',axis=1, inplace=True) ## 77% missing in cabin columns so we want to drop it
data.drop('PassengerId',axis=1,inplace=True)
data.drop('Name',axis=1.,inplace=True)
data.drop('Ticket',axis=1,inplace=True)
```

```python
data.head()
```

```python
data['Embarked'].unique()
```

```python
#Labelled Encoding(use mannual encoding)
data['Sex']=data['Sex'].map({'male':1,'female':0})
data['Embarked']=data['Embarked'].map({'S':1, 'C':2, 'Q':3})
```

```python
data.head()
```

```python
data.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
datacol=['Survived', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare','Embarked']
plt.figure(figsize=(10,10),facecolor='grey')
plotnumber=1
for col in datacol:
    if plotnumber <= 8:
        ax=plt.subplot(3,3,plotnumber)
        sns.boxplot(data[col])
        plt.xlabel(col,fontsize=15)
        plt.ylabel('value',fontsize=15)
    plotnumber +=1
plt.tight_layout()
plt.show()
```

```python
# Handling outliers using IQR method
from scipy import stats
columns=['Age', 'SibSp', 'Parch', 'Fare']
for col in columns:
    IQR=stats.iqr(data[col],interpolation='midpoint')
    Q1=data[col].quantile(0.25)
    Q3=data[col].quantile(0.75)
    lb=Q1 - 1.5*IQR
    ub=Q3 + 1.5*IQR

    print(f"\nColumn:{col}")
    print('IQR=',IQR)
    print('Q1=',Q1)
    print('Q3=',Q3)
    print('Lower Bound=',lb)
    print('upper Bound=',ub)
    print("outliers above upper bound:",len(data.loc[data[col]>ub]))
    print("outliers below lower bound:",len(data.loc[data[col]<lb]))

    # Replace outliers with median
    median_value=np.median(data[col])
    data.loc[(data[col] > ub)|(data[col] < lb),col]=median_value
    print(len(data.loc[(data[col]>ub)|(data[col]<lb)]))

```

```python
# Squareroot transformation
data['Age'] = np.sqrt(data['Age'])
data['Fare']=np.sqrt(data['Fare'])
```

```python
datacol=['Survived', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare','Embarked']
plt.figure(figsize=(10,10),facecolor='grey')
plotnumber=1
for col in datacol:
    if plotnumber <= 8:
        ax=plt.subplot(3,3,plotnumber)
        sns.boxplot(data[col])
        plt.xlabel(col,fontsize=15)
        plt.ylabel('value',fontsize=15)
    plotnumber +=1
plt.tight_layout()
plt.show()

```

```python
# Data Scalling
from sklearn.preprocessing import MinMaxScaler
ms=MinMaxScaler()
cols=['Pclass', 'Age', 'SibSp', 'Parch', 'Fare','Embarked']
d=data[cols]
df=ms.fit_transform(d)
df=pd.DataFrame(df,columns=cols)
```

```python
data = pd.concat([data[['Survived','Sex']],df],axis=1)
```

# Feature selection

```python
data.head()
```

```python
# separation of input output variable
x = data.drop('Survived', axis=1)
y=data.Survived
```

```python
x
```

```python
# train test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42)
```

```python
# Data Balancing 
from imblearn.over_sampling import SMOTE
smote = SMOTE() ## object creation
```

```python
y_train.value_counts()
```

```python
x_smote, y_smote = smote.fit_resample(x_train.astype('float'),y_train)
```

```python
y=np.array(y_smote)
```

```python
y_smote
```

```python
from collections import Counter
print("Actual Classes",Counter(y_train))
print("SMOTE Classes",Counter(y_smote))
```

# Model Building

```python
# Model creation
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_smote,y_smote)  ## training
```

```python
#Prediction
y_pred=lr.predict(x_test)
```

```python
y_pred
```

## Evaluation of Model

```python
from sklearn.metrics import confusion_matrix,accuracy_score,recall_score,precision_score,classification_report,f1_score
```

### Testing Data Accuarcy

```python
y_pred=lr.predict(x_test)
test_accuracy = accuracy_score(y_test, y_pred)
print("Testing accuracy is :", test_accuracy)
print()
clf_report = classification_report(y_test, y_test)
print("Classification report is\n", clf_report)
print()

conf_matrix = confusion_matrix(y_test, y_test)
print("Confusion matrix\n", conf_matrix)
```

## Training data accuracy

```python
y_pred_train =lr.predict(x_smote)

train_accuracy = accuracy_score(y_smote, y_pred_train)
print("Training accuracy is :", train_accuracy)
print()
clf_report = classification_report(y_smote, y_pred_train)
print("Classification report is\n", clf_report)
print()

conf_matrix = confusion_matrix(y_smote, y_pred_train)
print("Confusion matrix\n", conf_matrix)
```

## “The model achieved an accuracy of 76%, which means it correctly classified 76% of the total samples in the test dataset.”
